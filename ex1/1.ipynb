```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#2.1 plotting the data
data=np.loadtxt("ex1/data/ex1data1.txt",delimiter=",")
x=data[:,0]
y=data[:,1]
m=len(x)
plt.xlabel("Profit in $10,000s") 
plt.ylabel("'Population of City in 10,000s") 
plt.plot(y,x,"x",10)
plt.show()
```

```python
#2.2 gradient descent
x=np.concatenate((np.ones(m).reshape(m,1),data[:,0].reshape(m,1)),axis=1)
theta=np.zeros(2)
iterations=1500
alpha=0.01
def H(x,theta):
    return np.dot(x,theta)
def J(x,y,theta):
    return np.mean(np.square(H(x,theta)-y),axis=0)/2.0
print(J(x,y,theta))
def Update(x,y,theta):
    theta=theta-np.mean(alpha*(H(x,theta)-y).reshape((m,1))*x,axis=0)
    return theta
theta_vals=np.zeros([iterations,theta.shape[0]])
for i in range(iterations):
    theta=Update(x,y,theta)
    if(i%100==0):
        print('cost after ',i,' iterations: ',J(x,y,theta))
    theta_vals[i]=theta
```

```python
#2.4 visualizing J(theta)
plt.plot(theta_vals[:,0],theta_vals[:,1],'x')
plt.show()
```

```python
#3.1 feathre normalization
data=np.loadtxt("ex1/data/ex1data2.txt",delimiter=",")
for i in range(data.shape[1]):
    mean=np.mean(data[:,i])
    std=np.std(data[:,i])
    data[:,i]=(data[:,i]-mean)/std
x=data[:,:-1]
y=data[:,-1]
```

```python
#3.2 gradient descent
m=y.shape[0]
x=np.concatenate((np.ones(m).reshape(m,1),data[:,0].reshape(m,1)),axis=1)
theta=np.zeros(2)
iterations=50
alpha=0.1
def H(x,theta):
    return np.dot(x,theta)
def J(x,y,theta):
    return np.mean(np.square(H(x,theta)-y),axis=0)/2.0
def Update(x,y,theta):
    theta=theta-np.mean(alpha*(H(x,theta)-y).reshape((m,1))*x,axis=0)
    return theta
theta_vals=np.zeros([iterations,theta.shape[0]])
J_vals=np.zeros([iterations,1])
for i in range(iterations):
    theta=Update(x,y,theta)
    #if(i%100==0):
    #    print('cost after ',i,' iterations: ',J(x,y,theta))
    theta_vals[i]=theta
    J_vals[i]=J(x,y,theta)
plt.plot(np.arange(iterations),J_vals)
plt.show()
```
